```python
import torch
from ultralytics import YOLO
```

```python
# Checking CUDA availability
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device")
```

## Loading pretrained model

```python
# Load YOLOv8 model (pretrained weights)
model = YOLO("yolo11m.pt")
```

## Pretrained accuracy on train and validation set

```python
# checking metrics of pretrained model on train set
initial_train_metrics = model.val(data='/kaggle/input/weapon-detection-16-guns-augmented-approach-2/weapon-detection-16/data.yaml', split='train', device=device)
print(initial_train_metrics)
```

```python
# checking metrics of pretrained model on validation set
initial_val_metrics = model.val(data='/kaggle/input/weapon-detection-16-guns-augmented-approach-2/weapon-detection-16/data.yaml', split='val', device=device)
print(initial_val_metrics)
```

## Model Training

```python
# Train the model using GPU if available
results = model.train(data="/kaggle/input/weapon-detection-16-guns-augmented-approach-2/weapon-detection-16/data.yaml", epochs=150, imgsz=640, amp=False, patience=10, augment=True, device=device)
```

## Training and Validation Accuracy

```python
metrics_train = model.val(device=device, split="train")
print(metrics_train)
```

```python
# Validate the model
metrics_val = model.val(device=device, split="val")
print(metrics_val)
```
