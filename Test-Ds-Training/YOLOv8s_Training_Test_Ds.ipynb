```python
import torch
from ultralytics import YOLO
```

```python
# Checking CUDA availability
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device")
```

```python
# Load YOLOv8 model (pretrained weights)
model = YOLO("yolov8s.pt")
```

```python
# Train the model using GPU if available
results = model.train(data="/kaggle/input/test-ds/test_ds-1/data.yaml", epochs=100, imgsz=320, device=device) 
```

```python
# Validate the model
metrics = model.val(device=device)
print(metrics)
```
