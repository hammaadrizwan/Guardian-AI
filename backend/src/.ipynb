```python
import os
import pandas as pd
import shutil

csv_path = '_annotations.csv'
image_folder = './test'
destination_folder = 'other'


os.makedirs(destination_folder, exist_ok=True)

df = pd.read_csv(csv_path)

# Loop through the DataFrame
for index, row in df.iterrows():
    image_class = str(row['class']).strip()
    image_filename = str(row['filename']).strip()  # <- Strip spaces here

    if image_class != 'person':
        source_path = os.path.join(image_folder, image_filename)
        destination_path = os.path.join(destination_folder, image_filename)

        if os.path.exists(source_path):
            shutil.move(source_path, destination_path)
            print(f"Moved: {image_filename}")
        else:
            print(f"File not found: {image_filename} — checked path: {source_path}")


```
