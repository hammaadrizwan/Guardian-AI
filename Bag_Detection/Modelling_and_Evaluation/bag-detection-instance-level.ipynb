## Object Tracking Logic

```python
model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt").to(device) # Load the best trained model
print(f"Model loaded and pushed to {device}")
```

```python
tracker = DeepSort(max_age=30) # deep sort instance, max_age=30 keeps track of objects for 230 seconds
```

```python
# Load pretrained classification model
classifier = models.resnet50(pretrained=True).to(device)
classifier.eval()
```

```python
# Image transformation for resnet
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])
```

```python
# Store timestamps and embeddings of detected objects
unattended_objects = {}
```

```python
# Function to get object embedding (Move tensor computations to CUDA)
def get_object_embedding(image):
    image = transform(image).unsqueeze(0).to(device)  # Move tensor to GPU
    with torch.no_grad():
        embedding = classifier(image)
    return embedding.cpu()  # Move back to CPU for further processing
```

```python
# Function to check if a new object has replaced the old one
def is_new_object(old_embedding, new_embedding, threshold=0.9):
    similarity = torch.cosine_similarity(old_embedding, new_embedding)
    return similarity < threshold  # If similarity is low, it's a new object
```

```python
# Open the video stream
cap = cv2.VideoCapture("bag_detection.mp4")
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to torch tensor and move to CUDA if available
    frame_tensor = torch.from_numpy(frame).to(device)

    # Run YOLO detection on GPU
    results = model(frame_tensor)

    detections = []  # List for DeepSORT tracking

    for r in results:
        for box in r.boxes.data:
            x1, y1, x2, y2, conf, class_id = box.cpu().numpy()
            detections.append(([x1, y1, x2, y2], conf, class_id))

    # Perform Object Tracking
    tracked_objects = tracker.update_tracks(detections, frame=frame)

    current_time = time.time()
    for track in tracked_objects:
        track_id = track.track_id
        bbox = track.to_tlbr()  # Convert to (x1, y1, x2, y2) format

        x1, y1, x2, y2 = map(int, bbox)

        # Crop the detected object
        object_crop = frame[y1:y2, x1:x2]

        if object_crop.size == 0:  # Ignore invalid crops
            continue

        # Convert to PIL image for embedding calculation
        object_pil = Image.fromarray(cv2.cvtColor(object_crop, cv2.COLOR_BGR2RGB))
        new_embedding = get_object_embedding(object_pil)

        if track_id in unattended_objects:
            time_placed, old_embedding = unattended_objects[track_id]
            time_elapsed = current_time - time_placed

            # Check if object has been replaced
            if is_new_object(old_embedding, new_embedding):
                unattended_objects[track_id] = (current_time, new_embedding)  # Reset timer
            elif time_elapsed > 30:  # If the object remains for more than 30 sec
                cv2.putText(frame, "POTENTIAL THREAT!", (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
        else:
            unattended_objects[track_id] = (current_time, new_embedding)  # Store first-time object entry

        # Draw bounding box
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, f"ID {track_id}", (x1, y1 - 30), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    # Display the frame
    cv2.imshow("Frame", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
# Release video capture and close windows
cap.release()
cv2.destoryAllWindows()
```
